# Localitzador de zones d'interés utilitzant ORB

Primer de tot importem les llibreries neceissàries:

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

Importem la imatge i utilitzant funcions de la llibreria cv2 detectem els punts d'interés amb la funció detectAndCompute() i dibuixem els punts amb drawKeypoints(). Per mostrar els resulitats utilitzem imshow de la llibreria matplotlib.pyplot:

```python
img = cv2.imread('/home/marta/Desktop/GDSA/Mercat.JPG',0)
#linea que soluciona un bug de opencv
cv2.ocl.setUseOpenCL(False)
# Initiate STAR detector
orb = cv2.ORB_create()

# find the keypoints with ORB
kp = orb.detect(img,None)

# compute the descriptors with ORB
kp, des = orb.compute(img, kp)

# draw only keypoints location,not size and orientation
img2 = cv2.drawKeypoints(img,kp,None,color=(255,0,0), flags=0)
plt.imshow(img2),plt.show()
```
